```python
import numpy as np
import matplotlib.pyplot as plt
import math 

#SI unit
h =  6.62*10**-34      #planck constant
c =  3*10**8           #speed of light
k =  1.38*10**-23      #Boltzmann Constant JK^-1 

num = 100       #total numbers of particles

v =  np.linspace(10**0, 10**15.5, num, endpoint=True)


T =  5500
B =  2*h*v**3/c**2/(math.e**(h*v/k/T)-1)


#plt.xlim(10**15, 10**15.2)
#plt.ylim(10**-14, 10**-6)
plt.plot(v,B,".")
#plt.loglog(v,B,".")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Intensity Watts/Hz/m^2")

dB = np.random.normal(0.,10**-8.5, num)
B += dB
plt.plot(v,B)
#plt.loglog(v,B)

plt.show()
```
